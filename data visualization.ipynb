```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt     #import libraries
```

```python
plt.plot([1,2,3,4,5],[0,2,4,6,8],'--go',marker='v',markeredgecolor="red") # marker for green line and red triangle
plt.title('MY FIRST PLOT')  # TITLE GIVEN
plt.savefig('Myfirstimage.png',dpi=100)
plt.xlabel('x-axis')
plt.ylabel('y-axis') # labels for axis
plt.xticks([1,2,3,4,5,6,7])
plt.yticks([2,4,6,8,10,12]) #scaling the points or customization
plt.legend("x")
plt.show()  #basic line plot
```

```python
# BAR CHART
labels=['SWEETS','CHOCOLATES','TOYS']
values=[1,3,5]
bars=plt.bar(labels,values)
bars[0].set_hatch('/')# represents comparisons between discrete variables and categorical data.
bars[1].set_hatch('/')# sets pattern
bars[2].set_hatch('/')
plt.show()
```

```python
plt.scatter(x=[1,2,3,4],y=[0,2,4,6])
```

```python
x=[1,2,3,4,5]
y=[0,2,4,6,8]

hist=plt.hist(x,bins=y)# histograms used for frequency distribution for conntinuos variables and numerical data

plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
b_s=[100,150,170,158,134,200,157,200,146,100,123]
plt.hist(b_s)
plt.show() #to only get plot and ignore the details
```

```python
plt.hist(b_s,bins=[100,120,140,160,180,200],rwidth=0.75,color="purple")#relative width and bins added acc to range
plt.show()
```

```python
#histogram for 2 variables
bs_men=[120,230,200,134,135,167,189,128,158,123,156]
bs_women=[120,145,167,182,156,190,200,136,173,156,187]
plt.hist([bs_men,bs_women],bins=[120,140,160,180,200,220],color=['yellow','purple'],label=['men','women'])
plt.title("BLOOD SUGAR LEVELS")
plt.xlabel("BLOOD SUGAR VALUES")
plt.ylabel("NO OF PERSONS")
plt.legend()
plt.show()
```

```python
#pie chart

labels=['RENT','FOOD','CLOTHING','ELECTRONICS','AUTOMOBILE','MISCELLANEOUS']
values=[10000,2000,5000,3000,2000,5000]
plt.pie(values,labels=labels,autopct='%0.1f%%',shadow=True,explode=[0.1,0.5,0.3,0.2,0.5,0.3])
plt.axis("equal")
plt.savefig("MY PIE CHART.png")
plt.show()
```

```python
#box plot
dat=[np.random.normal(6,std,100)for std in range(1,2)]# 6 is median to see the spread of data
plt.boxplot(dat,vert=False,patch_artist=False)
plt.show()
```
